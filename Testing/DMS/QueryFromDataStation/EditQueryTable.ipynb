```python
df_ori, meta = pyreadstat.read_sav('AV291v3.sav')
```

```python
value_labels = meta.variable_value_labels
for c in df_col:
    if c in value_labels:
        if c != 'NIT_kcal' and c != 'medGRscore_TR':
            print(c)
            if c == 'n_smokingcat4':
                temp = {0.0: 'never', 1.0: 'former &gt;6mnd', 2.0: 'former &lt;6mnd', 3.0: 'current'}
                new_dict = dict([(value, key) for key, value in temp.items()]) 
                df_V1 = df_V1.replace({c: new_dict})
                
            elif c == 'N_ALCOHOL_CAT':
                temp = {0.0: 'None', 1.0: 'Low (women:&lt;=7; men:&lt;=14)', 2.0: 'High (women&gt;7; men&gt;14)'}
                new_dict = dict([(value, key) for key, value in temp.items()]) 
                df_V1 = df_V1.replace({c: new_dict})
                
            else:
                new_dict = dict([(value, key) for key, value in value_labels[c].items()]) 
                df_V1 = df_V1.replace({c: new_dict})
```
