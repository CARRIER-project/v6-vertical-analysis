```python
import pandas as pd
```

```python
df=pd.read_csv('cbs_test.csv')
df.columns
```

```python
GBAGeboorteJaar = []
GBAGeboorteMaand =[]
GBAGeboorteDag = []
GBAGeslacht = []
GBAPostcode = []
GBAHuisnummer = []
GBAToev = []
```

```python
dateOfBirth = list(df['date_of_birth'])
for i in dateOfBirth:
    GBAGeboorteJaar.append(str(i)[0:4])
    GBAGeboorteMaand.append(str(i)[4:6])
    GBAGeboorteDag.append(str(i)[6:8])
GBAGeslacht = list(df['sex'])
GBAPostcode = list(df['zipcode'])
GBAHuisnummer = list(df['housenum'])
add = list(df['zipcode'])
for i in add:
    GBAToev.append(i[-1])
```

```python
new_test_data = pd.DataFrame.from_records([GBAGeboorteJaar,GBAGeboorteMaand,GBAGeboorteDag,GBAGeslacht,\
                           GBAPostcode,GBAHuisnummer,GBAToev]).transpose()
new_test_data.columns = ['GBAGeboorteJaar','GBAGeboorteMaand','GBAGeboorteDag','GBAGeslacht',\
                           'GBAPostcode','GBAHuisnummer','GBAToev']
new_test_data
```

```python
##### Generate simulated CBS Data file #####
```

```python
fileList = ['Vektis2011.csv','Vektis2012.csv','Vektis2013.csv','Vektis2014.csv','Vektis2015.csv',\
        'Vektis2016.csv','Vektis2017.csv']

def getVar(file,year):
    df = pd.read_csv(file,delimiter=';')[1:]
    df = df.reset_index(drop=True)
    
    file_col = list(df.columns)
    req_col = ['LEEFTIJDSKLASSE', 'GESLACHT', 'AANTAL_BSN',\
                        'KOSTEN_MEDISCH_SPECIALISTISCHE_ZORG', 'KOSTEN_FARMACIE',\
                        'KOSTEN_SPECIALISTISCHE_GGZ', 'KOSTEN_HUISARTS_INSCHRIJFTARIEF',\
                        'KOSTEN_HUISARTS_CONSULT', 'KOSTEN_HUISARTS_MDZ',\
                        'KOSTEN_HUISARTS_OVERIG', 'KOSTEN_HULPMIDDELEN', 'KOSTEN_MONDZORG',\
                        'KOSTEN_PARAMEDISCHE_ZORG_FYSIOTHERAPIE',\
                        'KOSTEN_PARAMEDISCHE_ZORG_OVERIG', 'KOSTEN_ZIEKENVERVOER_ZITTEND',\
                        'KOSTEN_ZIEKENVERVOER_LIGGEND', 'KOSTEN_KRAAMZORG',\
                        'KOSTEN_VERLOSKUNDIGE_ZORG', 'KOSTEN_GENERALISTISCHE_BASIS_GGZ',\
                        'KOSTEN_LANGDURIGE_GGZ', 'KOSTEN_GRENSOVERSCHRIJDENDE_ZORG',\
                        'KOSTEN_EERSTELIJNS_ONDERSTEUNING','KOSTEN_GERIATRISCHE_REVALIDATIEZORG',\
                        'KOSTEN_VERPLEGING_EN_VERZORGING', 'KOSTEN_OVERIG']
    new_file_col = []
    new_file_col_name = []
    for i in req_col:
        if i in file_col:
            new_file_col.append(i)
            if i != 'LEEFTIJDSKLASSE' and i !='AANTAL_BSN' and i!= 'GESLACHT':
                new_file_col_name.append(i+'_'+year)
            else:
                new_file_col_name.append(i)

    df = df[new_file_col]
    df.columns = new_file_col_name
    df = df[0:5001]
    
    return df,new_file_col_name
```

```python
all_df = {}
new_file_col_name = {}
for file in fileList:
    year = file[6:-4]
    all_df[year],new_file_col_name[year] = getVar(file,year)
```

```python
new_cbs_data_df=pd.concat([new_test_data,all_df['2011'],
           all_df['2012'][new_file_col_name['2012'][2:]],
           all_df['2013'][new_file_col_name['2013'][2:]],
           all_df['2014'][new_file_col_name['2014'][2:]],
           all_df['2015'][new_file_col_name['2015'][2:]],
           all_df['2016'][new_file_col_name['2016'][2:]],
           all_df['2017'][new_file_col_name['2017'][2:]]], axis=1, join='inner')[:]
new_cbs_data_df
```

```python
new_cbs_data_df.to_csv('new_cbs_test_enc.csv',index=None)
```
